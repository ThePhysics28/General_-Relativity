```python
import sympy as sp

# Sembolik değişkenleri tanımla
r, theta, V_1, V_2 = sp.symbols('r theta V_1 V_2')

# Koordinat Sistemi 1 (Kartezyen Koordinatlar)
A_1 = r * sp.cos(theta)
A_2 = r * sp.sin(theta)

# Koordinat Sistemi 2 (Polar Koordinatlar)
B_1 = r
B_2 = theta

# Dönüşüm matrisi için gerekli türevler
A_1_1 = sp.diff(A_1, B_1)  # ∂(x) / ∂r
A_1_2 = sp.diff(A_1, B_2)  # ∂(x) / ∂θ
A_2_1 = sp.diff(A_2, B_1)  # ∂(y) / ∂r
A_2_2 = sp.diff(A_2, B_2)  # ∂(y) / ∂θ

# Dönüşüm matrisini tanımla
T = sp.Matrix([[A_1_1, A_1_2],
               [A_2_1, A_2_2]])

# Matrisin tersini al (T^(-1))
T_inv = T.inv()

# Eski koordinatlardaki vektörü tanımla
V = sp.Matrix([V_1, V_2])

# Yeni koordinat sistemindeki vektörü hesapla
V_new = T_inv * V

# Sonuçları yazdır
print("Transformation Matrix T:")
sp.pprint(T)

print("\nInverse Transformation Matrix T^(-1):")
sp.pprint(T_inv)

print("\nVector in New Coordinate System:")
sp.pprint(V_new)


```
